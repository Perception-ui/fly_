```python
# Selecting only numerical features for correlation analysis
numerical_features = data[['num_passengers', 'purchase_lead', 'length_of_stay', 'flight_hour', 'flight_duration']]

# Calculate the correlation matrix
correlation_matrix = numerical_features.corr()
# Display the correlation matrix in numeric form
print("Correlation Matrix:")
print(correlation_matrix)
```

Values close to 1 or -1 indicate a strong positive or negative correlation, respectively.
Values close to 0 indicate a weak or no correlation.
In the dataset, most correlations appear to be weak, indicating that the features are not highly correlated. This is good as it suggests that there is not a significant issue with multicollinearity.

```python
# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

Identifying outliers

```python
# Calculate the IQR for each numerical feature
Q1 = numerical_features.quantile(0.25)
Q3 = numerical_features.quantile(0.75)
IQR = Q3 - Q1

# Identify potential outliers
outliers = ((numerical_features < (Q1 - 1.5 * IQR)) | (numerical_features > (Q3 + 1.5 * IQR))).any(axis=1)

# Display the rows with potential outliers
print("Rows with potential outliers:")
print(data[outliers])
# Create box plots to visualize outliers
plt.figure(figsize=(15, 8))
sns.boxplot(data=numerical_features)
plt.title('Box Plots for Numerical Features')
plt.show()
```

Outliers are identified by higher numbers of passengers, longer purchase lead times, longer lengths of stay, and no significant differences in flight hour and duration. These outliers may be due to group bookings, advanced planning, or longer durations for specific travel purposes. The hour of flight departure and total flight duration are not significant factors.

## EDA 

### Univariate Analysis 

#### What is the distribution of the target variable (booking_complete)?

```python
# Create a count plot for the target variable
plt.figure(figsize=(8, 6))
sns.countplot(x='booking_complete', data=data)
plt.title('Distribution of Booking Completion')
plt.xlabel('Booking Completion (1: Completed, 0: Not Completed)')
plt.ylabel('Count')
plt.show()
```

The count plot shows an imbalance in the distribution of the target variable.Imbalanced datasets can pose challenges for machine learning models. In this case,  there is substantially more incompleted bookings than completed bookings a model might become biased toward predicting the majority class.
It can be addressed by Techniques such as oversampling the minority class, undersampling the majority class, whuch can help balance the dataset.

#### 2.What is the distribution of the numerical features (num_passengers, purchase_lead, length_of_stay, flight_hour, flight_duration)?

```python
# Selecting only numerical features for distribution analysis
numerical_features = data[['num_passengers', 'purchase_lead', 'length_of_stay', 'flight_hour', 'flight_duration']]

# Create subplots for each numerical feature
plt.figure(figsize=(15, 10))
for i, column in enumerate(numerical_features.columns, 1):
    plt.subplot(2, 3, i)
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Bookings with a smaller number of passengers are more frequent, indicating a preference for smaller groups 


A significant number of bookings are made on the day of or very close to the travel date (purchase_lead = 0), suggesting a tendency for last-minute bookings.

Bookings with a length of stay around 0 days are more common


Flights scheduled between 5-15 hours (hours 5 to 15) have a higher frequency of bookings, suggesting a preferred time range for travel.

Flights with a duration around 9 hours have the highest booking frequency,

#### 3.How are the binary preference variables (wants_extra_baggage, wants_preferred_seat, wants_in_flight_meals) distributed?

```python
# Selecting only binary preference variables for distribution analysis
binary_preferences = data[['wants_extra_baggage', 'wants_preferred_seat', 'wants_in_flight_meals']]

# Create subplots for each binary preference variable
plt.figure(figsize=(15, 5))
for i, column in enumerate(binary_preferences.columns, 1):
    plt.subplot(1, 3, i)
    sns.countplot(x=column, data=data)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

The majority of customers do not prioritize preferred seats or in-flight meals, suggesting that their preferences may not significantly impact booking completion.For wants_extra_baggage, it might be valuable to explore how this preference correlates with the completion of bookings. Customers who request extra baggage might have different behaviors or completion rates.

### Bivariate Analysis:

#### 1.How does the completion of bookings vary with the number of passengers (num_passengers)?

```python
# Convert num_passengers to categorical
data['num_passengers'] = data['num_passengers'].astype('category')

# Create a count plot for num_passengers with hue based on booking completion
plt.figure(figsize=(10, 6))
sns.countplot(x='num_passengers', hue='booking_complete', data=data, hue_order=[0, 1])
plt.title('Booking Completion by Number of Passengers')
plt.xlabel('Number of Passengers')
plt.ylabel('Count')
plt.legend(title='Booking Completion', labels=['Not Completed', 'Completed'])
plt.show()
```

The majority of bookings, regardless of the number of passengers, are not completed.The completion of the booking is not influenced by the number of passengers 

#### 2.Is there a relationship between the completion of bookings and the lead time between booking and travel (purchase_lead)?

```python
# Create histograms for purchase_lead based on booking completion
plt.figure(figsize=(12, 6))

# Histogram for not completed bookings
sns.histplot(data[data['booking_complete'] == 0]['purchase_lead'], bins=30, color='skyblue', label='Not Completed', kde=False)

# Histogram for completed bookings
sns.histplot(data[data['booking_complete'] == 1]['purchase_lead'], bins=30, color='salmon', label='Completed', kde=False)

plt.title('Distribution of Purchase Lead Time by Booking Completion')
plt.xlabel('Purchase Lead Time (days)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

the histogram indicates that the "Not Completed" category has a higher frequency in certain lead time ranges compared to the "Completed" category, it suggests that shorter lead times may be associated with a higher likelihood of bookings not being completed. 

#### 3.What is the impact of the length of stay (length_of_stay) on booking completion?

```python
# Create histograms for length_of_stay based on booking completion
plt.figure(figsize=(12, 6))

# Histogram for not completed bookings
sns.histplot(data[data['booking_complete'] == 0]['length_of_stay'], bins=30, color='skyblue', label='Not Completed', kde=False)

# Histogram for completed bookings
sns.histplot(data[data['booking_complete'] == 1]['length_of_stay'], bins=30, color='salmon', label='Completed', kde=False)

plt.title('Distribution of Length of Stay by Booking Completion')
plt.xlabel('Length of Stay (days)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

It indicates that the length_of_stay alone may not be a strong predictor of booking completion.

#### 4.Are there specific hours of the day (flight_hour) when bookings are more likely to be completed?

```python
# Create a count plot for flight_hour with hue based on booking completion
plt.figure(figsize=(12, 6))
sns.countplot(x='flight_hour', hue='booking_complete', data=data, palette='Set2')
plt.title('Booking Completion by Hour of Flight Departure')
plt.xlabel('Hour of Flight Departure')
plt.ylabel('Count')
plt.legend(title='Booking Completion', labels=['Not Completed', 'Completed'])
plt.show()
```

The period between 1 and 15 might be considered peak booking hours,The count plot for flight_hour indicates that there is a higher number of not completed bookings across different hours of the day. This suggests that the hour of flight departure alone might not be a strong predictor of booking completion

#### 5.Does the preference for extra baggage (wants_extra_baggage) influence booking completion?

```python
# Create a stacked bar plot for wants_extra_baggage and booking completion
plt.figure(figsize=(8, 6))
sns.countplot(x='wants_extra_baggage', hue='booking_complete', data=data, palette='pastel', edgecolor='black')
plt.title('Booking Completion by Extra Baggage Preference')
plt.xlabel('Wants Extra Baggage')
plt.ylabel('Count')
plt.legend(title='Booking Completion', labels=['Not Completed', 'Completed'])
plt.xticks([0, 1], ['No', 'Yes'])  # Optional: Adjust x-axis labels for better readability
plt.show()
```

The observation that "Not Completed" has the highest count in both "No" and "Yes" categories of the wants_extra_baggage variable suggests that the completion rate is generally lower, regardless of whether customers express a preference for extra baggage.
